## Taking a deeper look at True Shooting Percentage (TS%)

```python
# Showing the histogram for ts_pct (true shooting percent)

%matplotlib inline
histogram = nba_stats['ts_pct'].hist()
```

Here you a have a histogram of the ts_pct (True shooting percentage). True shooting percentage (TS%) is a measure of shooting efficiency that takes into account field goals, 3-point field goals, and free throws. In layman's terms, it helps calculate how efficient a player is truly shooting the basketball. This statistic has revolutionized the way that teams and players view scoring. As we can see above, most of the players in the NBA have a TS% around the 50% range.

```python
# Filtering the data to get only players who scored more than 22 pts per game

as_ppgs = nba_stats[nba_stats['pts'] >= 22]
```

Typically an NBA all-star averages more than 22 points per game. Using the above code, I am filtering out all of the players who averaged less than 22 points per game. Now by no means is the a perfect filter because there are all-star players who average less, but this is the best I can do.

```python
%matplotlib inline
histogram = as_ppgs['ts_pct'].hist()
```

As you can see, the average of TS% of players who are more than likley all-stars, now jumps to 57%, which is a substantial increase from the average player. This means that all-stars are shooting the ball at a more efficient rate than that of an average player.

## Finding the most average NBA player

For this we will be trying to find the player that has the most average stats in the NBA. Earlier we saw that an average NBA player averages 8.2 points, 3.6 rebounds, and 1.8 assists a game. Lets see if we can find a player that matches that category.

```python
# Getting a player with less than 8.2 points per game

avg_player = nba_stats[nba_stats['pts'] <= 8.2]
```

```python
# Getting a player with less than 3.6 rebounds per game

avg_player2 = avg_player[avg_player['reb'] <= 3.6]
```

```python
# Getting a player with less than 1.8 assists per game

avg_player3 = avg_player2[avg_player2['ast'] <= 1.8]
```

```python
# Here are the top 10 players with the closest averages to the most average NBA players

avg_player3.nlargest(10, ['pts', 'reb', 'ast'])
```

As we can see from the table above, these are the most average NBA players in the database when looking at points, rebounds and assits. By looking at this we can see that Bruce Bowen is truly the closest player we have to the most average NBA player. He averaged 8.2 points, 3.5 rebounds and 1.5 assits per game in the 04-05 season. He also had a TS% of 52%, which we established earlier as very close to the average of 50%.
